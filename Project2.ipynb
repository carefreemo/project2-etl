```python
# 1. Import libraries

from sqlalchemy import create_engine as ce
import pandas as pd
import numpy as np
```

```python
# 2. Import and read the National University Rankings csv file

Universities = "NationalUniversitiesRankingsFinal.csv"

Universities = pd.read_csv (Universities)
Universities
```

```python
# 3. Use the head function to refer school rankings (we are maininly looking for top 10 unis and to review the data)

Universities.head(15)
```

```python
# 4. Use the drop function to drop any university that is not a top 10 school

Universities.drop(Universities.index[10:231], inplace = True)
Universities
```

```python
# 5. Use the Data Frame function to create a data frame for the new University data set

UniversitiesClean = pd.DataFrame(Universities[["Name","Location","Rank"]])
UniversitiesClean
```

```python
# 6. Use the rename function to rename the 'Name' column to 'college'

UniversitiesClean.rename(columns={'Name': 'college'}, inplace=True)
UniversitiesClean
```

```python
#7. Import and Read the Players csv file 

Players = "PlayersFinal.csv"

Players = pd.read_csv (Players)
Players
```

```python
# 8. Use the DataFrame function to create a data frame for the Players file

PlayersClean = pd.DataFrame(Players[['_id', 'name', 'birthDate', 'birthPlace', 'college', 'draft_pick',
       'draft_round', 'draft_team', 'draft_year', 'highSchool', 'position', 'shoots', 'weight']])
PlayersClean
```

```python
# 9. Use the dropna function to drop any row that has NAn in the 'college' column

PlayersClean.dropna(subset=['college'])
```

```python
# 10. Use the str.contains function to find look iny  the 'college' column and replace any cell that contains 
#     'Stanford University' with 'Standford University'

PlayersClean[PlayersClean['college'].str.contains('Stanford University', na=False)]

```

```python
# 11. Use the str function to find look iny  the 'college' column and replace any cell that contains 
#     'Duke University' with 'Duke University'


PlayersClean[PlayersClean['college'].str.contains('Duke University', na=False)]

```

```python
# 12. Use the str function to find look iny  the 'college' column and replace any cell that contains 
#     'University of Pennsylvania' with 'University of Pennsylvania'

PlayersClean[PlayersClean['college'].str.contains('University of Pennsylvania', na=False)]
```

```python
PlayersClean.columns
```

```python
# 13. Use the merge function to perform a right join (PlayersClean (left) and UniversitiesClean (right))

Combineddf=pd.merge(PlayersClean,UniversitiesClean,on="college", how = 'right')
Combineddf.head()
```

```python
# Connect to PosgreSQL
pg_engine = ce("postgresql://postgres:postgres@localhost:5432/NBA_DB")
```

```python
# Create University Table in SQL

UniversitiesClean.to_sql(name='UniversitiesTable', con=pg_engine, if_exists='append', index=False)


```

```python
# Create Players Table in SQL

PlayersClean.to_sql(name='PlayersTable', con=pg_engine, if_exists='append', index=False)
```
